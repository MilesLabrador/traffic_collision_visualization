```python
pd_Table_new = pd_Table.with_column("street_type", pd_Table.apply(abbreviate, "street_type"))
collisions_withcombo = pd_Table_new.with_column('full_name', pd_Table_new.apply(combineStrings, 'street_name', 'street_type'))
collisions = collisions_withcombo.drop('violation_section', 'charge_desc','violation_type')
collisions2 = collisions.group('street_name', sum).drop('da')
display(collisions)
collisions.to_csv('collisions_clean.csv')
```
